```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

new_d['GDP'] = new_d['GDP'].str.replace(',', '')

# Convert the 'y' variable to a numeric type
y = new_d['GDP']
y = [float(i) for i in y]

model = LinearRegression()
X = new_d[['YEAR']]
model.fit(X, y)
plt.plot(new_d['YEAR'],new_d['GDP'])
plt.xlim(2010,2025)
plt.ylim(16,0)
plt.show()
```

```python
new_data = np.array([[2025], [2026], [2027],[2028],[2029]])
new_predictions = model.predict(new_data)

# Print the new predictions
print("New data predictions:", new_predictions)
```
